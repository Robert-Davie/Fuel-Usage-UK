```python
import pandas as pd
```

```python
xls = pd.ExcelFile("Subnational_total_final_consumption_2005_2022.xlsx")
```

```python
sheets = {}
for sheet in range(2005, 2023):
    this_df = pd.read_excel(xls, str(sheet))
    sheets[sheet] = this_df
```

```python
sheets_2 = {}
for key in sheets.keys():
    letter_to_country = {
        "E": "England",
        "W": "Wales",
        "S": "Scotland",
        "N": "Nothern Ireland",
    }
    cp = sheets[key]
    cp.columns = cp.iloc[4]
    cp = cp.iloc[29:].copy()
    cp["Country"] = cp["Code"].str[0].map(letter_to_country)
    cp = cp.drop(["Code", "Notes"], axis=1)
    cp = cp.rename(columns={"Country or region": "Region"})
    cp = cp.reset_index().drop("index", axis=1)
    cp["Year"] = key
    sheets_2[key] = cp
```

```python
for sheet_name, sheet in sheets_2.items():
    sheet.to_csv(f"outputs/{sheet_name}.csv")
```
